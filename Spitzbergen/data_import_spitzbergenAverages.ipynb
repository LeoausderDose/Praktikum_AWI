```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the desired length for zero padding
desired_length = 1000000  # Adjust this to your specific needs

tagesgang_grad_mean = []

for i in range(0, 12):
    tagesgang_grad = []
    for y in range(len(years)):
        data = years[y][i]['T2 [°C] (average, Thermometer)'] - np.mean(dataframes[i]['T2 [°C] (average, Thermometer)'])

        # Zero padding: Extend the length of data to the desired length
        data = np.pad(data, (0, desired_length - len(data)), 'constant')

        # Apply a windowing function (e.g., Hanning window) to the data
        window = np.hanning(len(data))
        data = data * window

        fft = np.fft.fft(data)
        t = np.arange(len(data))
        freq = np.fft.fftfreq(t.shape[-1], d=1 / (60 * 24))

        amplitude = np.abs(fft) * desired_length / len(data)
        temp_index = np.where(freq == find_nearest(freq, 1))
        tagesgang_grad.append(amplitude[temp_index])

        # You can plot the frequency spectrum if needed
        # plt.plot(freq, amplitude)
        # plt.xlim(-2, 2)
        # plt.ylim(0, 0.1)

    plt.plot(tagesgang_grad, label=str(i))
    tagesgang_grad_mean.append(np.mean(tagesgang_grad))

plt.legend()
tagesgang_grad_mean
```
