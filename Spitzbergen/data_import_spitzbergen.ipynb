```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample frequency-domain data (for example, a complex sinusoidal signal)
n_samples = 1000
t = np.linspace(0, 1, n_samples, endpoint=False)  # Time values
freq = 5  # Frequency of the signal
amplitude = 1  # Amplitude of the signal
signal_freq_domain = amplitude * np.exp(1j * 2 * np.pi * freq * t)

# Perform the inverse FFT to obtain the time-domain signal
time_domain_signal = np.fft.ifft(signal_freq_domain)

# Plot the time-domain signal
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.plot(t, signal_freq_domain.real, label='Real part')
plt.plot(t, signal_freq_domain.imag, label='Imaginary part')
plt.xlabel('Time')
plt.title('Frequency-Domain Signal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, time_domain_signal.real, label='Real part')
plt.plot(t, time_domain_signal.imag, label='Imaginary part')
plt.xlabel('Time')
plt.title('Time-Domain Signal')
plt.legend()

plt.tight_layout()
plt.show()
```
